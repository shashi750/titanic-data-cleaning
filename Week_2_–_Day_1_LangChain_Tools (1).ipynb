```python
from langchain.agents import Tool

def square(x: str) -> str:
    return str(int(x) / 10)

tool = Tool(
    name="Divide",
    func=square,
    description="divide a number by 10"
)

print(tool.run("100"))  # Output: 50
```

```python
!pip install -q langchain transformers

from langchain.llms import HuggingFaceHub
from langchain.agents import initialize_agent, Tool, AgentType
from langchain.prompts import PromptTemplate
import os

# Replace with your Hugging Face token
os.environ["HUGGINGFACEHUB_API_TOKEN"] = "your_token_here"

# Setup LLM
llm = HuggingFaceHub(
    repo_id="google/flan-t5-large",
    model_kwargs={"temperature": 0.5, "max_length": 100}
)

# Simple tool: multiply number by 10
def multiply(x: str) -> str:
    return str(int(x) * 10)

tools = [
    Tool(
        name="Multiplier",
        func=multiply,
        description="Multiplies a number by 10"
    )
]

# Create agent using Hugging Face model
agent = initialize_agent(
    tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True
)

# Run the agent
agent.run("What is 7 multiplied by 10?")
```

```python
# Step 1: Install required packages
!pip install -q transformers langchain gradio huggingface_hub

# Step 2: Import required libraries
from transformers import pipeline
from langchain.llms import HuggingFacePipeline
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
import gradio as gr

# Step 3: Set up HuggingFace model pipeline
generator = pipeline(
    "text2text-generation",
    model="google/flan-t5-base",
    max_new_tokens=100,
    temperature=0.7,
    repetition_penalty=1.2
)

# Step 4: Wrap pipeline in LangChain
llm = HuggingFacePipeline(pipeline=generator)

# Step 5: Create PromptTemplate and Chain
template = "Write a {tone} paragraph about {topic}."
prompt = PromptTemplate.from_template(template)
chain = LLMChain(prompt=prompt, llm=llm)

# Step 6: Define Gradio interface function
def generate_paragraph(tone, topic):
    return chain.run({"tone": tone, "topic": topic})

# Step 7: Gradio UI setup
iface = gr.Interface(
    fn=generate_paragraph,
    inputs=[
        gr.Textbox(label="Tone (e.g., excited, formal)"),
        gr.Textbox(label="Topic (e.g., Artificial Intelligence)")
    ],
    outputs="text",
    title="🧠 LLM Paragraph Generator",
    description="Enter a tone and a topic. Get a smart paragraph generated by FLAN-T5 via LangChain + Hugging Face!"
)

# Step 8: Launch the app
iface.launch(share=True)  # share=True will give you a public link
```

```python
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

# Load HF model (T5 for text2text-generation)
generator = pipeline("text2text-generation", model="google/flan-t5-base", max_new_tokens=100)
llm = HuggingFacePipeline(pipeline=generator)

# Setup memory
memory = ConversationBufferMemory()

# Create conversational chain
conversation = ConversationChain(llm=llm, memory=memory, verbose=True)
```

```python
from langchain.memory import ConversationBufferMemory, ConversationSummaryMemory
from langchain.chains import ConversationChain
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

# Load model
gen = pipeline("text2text-generation", model="google/flan-t5-base", max_new_tokens=100)
llm = HuggingFacePipeline(pipeline=gen)

# Option 1: Buffer Memory
buffer_memory = ConversationBufferMemory()
buffer_chain = ConversationChain(llm=llm, memory=buffer_memory, verbose=True)

# Option 2: Summary Memory
from langchain.memory import ConversationSummaryMemory
summary_memory = ConversationSummaryMemory(llm=llm)
summary_chain = ConversationChain(llm=llm, memory=summary_memory, verbose=True)

# Sample conversation
print("Buffer Memory:")
print(buffer_chain.run("Hi, my name is Raj."))
print(buffer_chain.run("What did I tell you about myself?"))

print("\nSummary Memory:")
print(summary_chain.run("Hi, my name is Raj."))
print(summary_chain.run("What did I tell you about myself?"))
```

```python
from transformers import pipeline
from langchain.llms import HuggingFacePipeline
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory
import gradio as gr

# Load the HuggingFace text2text model
text2text_gen = pipeline("text2text-generation", model="google/flan-t5-base", max_new_tokens=100)
llm = HuggingFacePipeline(pipeline=text2text_gen)

# Initialize memory
memory = ConversationBufferMemory()

# Initialize the chain
conversation = ConversationChain(llm=llm, memory=memory, verbose=True)

# Gradio UI
def chat_with_ai(user_input):
    return conversation.run(user_input)

chatbot_ui = gr.Interface(fn=chat_with_ai,
                          inputs=gr.Textbox(label="You"),
                          outputs=gr.Textbox(label="🤖 AI"),
                          title="🧠 Chat with Memory")

chatbot_ui.launch(share=True)  # Shareable link for Colab
```
