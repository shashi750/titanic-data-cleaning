```python
import pandas as pd  # Import the pandas library for data manipulation

# Step 1: Load the Titanic dataset from a public GitHub URL
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

# Display the first 5 rows of the dataset
print(df.head())

# Step 2: Check for missing (null) values in each column
print(df.isnull().sum())

# Step 3: Calculate the average age (mean) from the 'Age' column
mean_age = df["Age"].mean()

# Replace missing values in the 'Age' column with the calculated mean
df["Age"] = df["Age"].fillna(mean_age)

# Step 4: Import libraries for data visualization
import seaborn as sns  # For plotting heatmaps and other advanced visualizations
import matplotlib.pyplot as plt  # For general plotting

# Step 5: Draw a correlation heatmap to understand relationships between numeric columns
plt.figure(figsize=(8,6))  # Set the size of the plot
corr = df.select_dtypes(include='number').corr()  # Select only numeric columns
sns.heatmap(corr, annot=True, cmap="Blues")  # Create heatmap
plt.show()  # Display the plot


# Step 6: Save the cleaned dataset as a new CSV file (without the index column)
df.to_csv("titanic_cleaned_v2.csv", index=False)



```
